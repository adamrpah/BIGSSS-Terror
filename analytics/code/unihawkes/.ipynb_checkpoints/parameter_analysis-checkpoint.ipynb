```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
dfs = []
for country in ['Afghanistan', 'Colombia', 'Iraq']:
    for gdir in glob.glob('../../results/unihawkes/%s/*/' % country):
        temp = pd.read_csv(gdir + 'averaged_parameters.csv', index_col=0)
        temp['gname'] = gdir.split('/')[-2]
        dfs.append(temp)
df = pd.concat(dfs)
df = df[df.gname != 'Unknown']
df.sort_values('gname', inplace=True)
```

```python
for param in ['mu', 'alpha', 'beta']:
    ax = df.plot(kind='bar', x='gname', y=param, color='black')
    ax.set_ylabel(param)
    ax.legend_.remove()
    ax.set_xlabel('')
    plt.savefig('%s.eps' % param)
```
